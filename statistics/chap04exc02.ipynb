Import the libraries needed to plot the PMF and CDF of 1000 numbers generated by `random.random`.

```python
import thinkstats2
import thinkplot
import random

%matplotlib inline
```

Generate a list of 1000 random numbers using `random.random`.

```python
s = []
for i in range(1000):
    s.append(random.random())
print(s[:5])
```

Create a PMF of the 1000 random numbers.

```python
pmf = thinkstats2.Pmf(s)
thinkplot.Pmf(pmf)
thinkplot.Show(xlabel='Random number', ylabel='Probability')
```

Create a CDF of the random numbers.

```python
cdf = thinkstats2.Cdf(s)
thinkplot.Cdf(cdf)
thinkplot.Show(xlabel='Random number', ylabel='CDF')
```

The PMF and CDF both indicate a uniform distribution. The PMF shows each number has probability of 0.001 of occuring which is 1/1000. The CDF shows a linear line from 0 to 1, which represents a uniform distribution.
